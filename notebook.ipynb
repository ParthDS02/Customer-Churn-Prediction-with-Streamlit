```python
y = df[["Churn"]]
x = df[["Age","Gender","Tenure","MonthlyCharges"]]
```

```python
x #1 for Female and 0 for Male for numeric value
```

```python
x["Gender"] = x["Gender"].apply(lambda x: 1 if x == "Female" else 0)
```

```python
y
```

```python
x.head()
```

```python
type(x["Gender"][0])
```

```python
y["Churn"] = y["Churn"].apply(lambda x: 1 if x == "Yes" else 0)
```

```python
type(y["Churn"][0])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2) #20% of data for test and rest (80%) for train
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
import joblib
joblib.dump(scaler, "scaler.pkl")
```

```python
x_test = scaler.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.metrics import accuracy_score
def modelperformance(predication):
    print("Accuracy score on model is {}".format(accuracy_score(y_test,predication)))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression()
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
log_model.fit(x_train,y_train)
```

```python
y_pred = log_model.predict(x_test)
```

```python
modelperformance(y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
param_grid = {"n_neighbors": [3, 5, 7, 9],
              "weights" : ["uniform", "distance"]
}
```

```python
# Define the K-Nearest Neighbors model
knn = KNeighborsClassifier()
```

```python
gridkn = GridSearchCV(estimator=knn, param_grid=param_grid, cv=5)
```

```python
gridkn.fit(x_train, y_train)
```

```python
gridkn.best_params_
```

```python
y_pred = gridkn.predict(x_test)
```

```python
modelperformance(y_pred)
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC()
```

```python
param_grid = {"C": [0.01,0.1,0.5,1],
              "kernel" : ["linear", "rbf","poly"],
              }
```

```python
gridsvc = GridSearchCV(svm, param_grid, cv = 5)
```

```python
gridsvc.fit(x_train,y_train)
```

```python
gridsvc.best_params_
```

```python
y_pred = gridsvc.predict(x_test)
```

```python
modelperformance(y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
param_grid = {
    "criterion" : ["gini", "entropy"],
    "splitter" : ["best", "random"],
    "max_depth" : ["None", 10,20,30],
    "min_samples_split" : [2,5,10],
    "min_samples_leaf" : [1,2,4]
}
```

```python
grid_tree = GridSearchCV(DecisionTreeClassifier(),param_grid, cv = 5)
```

```python
grid_tree.fit(x_train,y_train)
```

```python
grid_tree.best_params_
```

```python
y_pred = grid_tree.predict(x_test)
```

```python
y_pred
```

```python
modelperformance(y_pred)
```
